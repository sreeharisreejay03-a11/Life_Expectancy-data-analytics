```python
df['GDP']=df['GDP'].astype(float)
```

```python
df['GDP'] = df.groupby('Country')['GDP'].transform(lambda x: x.fillna(x.mean()))
```

```python
df['percentage expenditure']=df['percentage expenditure'].astype(float)
```

```python
df['percentage expenditure'] = df.groupby('Country')['percentage expenditure'].transform(
    lambda x: x.fillna(x.mean()))
```

```python
df['Income composition of resources']=df['Income composition of resources'].ffill()
```

```python
df['Schooling']=df['Schooling'].astype(float)
```

```python
df['Schooling'] = df.groupby('Country')['Schooling'].transform(lambda x: x.fillna(x.mean()))
```

```python
df['Year']=df['Year'].astype(int)
```

```python
yearly = df.groupby('Year')['Life expectancy'].mean().reset_index(name='Life expectancy')
yearly
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10,5))

plt.plot(yearly['Year'], yearly['Life expectancy'])

plt.title("Average Life Expectancy Over Years")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")

plt.grid(True)

plt.show()
```

Conclusion: Despite a significant dip between 2001 and 2003, the overall global trend is positive, showing a steady recovery and growth from 2007 onwards. By 2015, average life expectancy reached its highest point in the dataset (above 71 years).

```python

top10 = df.groupby('Country')['Life expectancy'] .mean().sort_values(ascending=False).tail(10)
top10
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Life expectancy'],bins=20, color="steelblue", edgecolor="white")
plt.title("Distribution of Life Expectancy")
plt.xlabel("Life Expectancy")
plt.ylabel("Frequency")
plt.grid()

plt.show()
```

Conclusion: The data is negatively skewed (left-skewed), with a high concentration of observations between 70 and 80 years. This indicates that while most recorded populations have a relatively high life expectancy, there is a long "tail" of regions with significantly lower outcomes (below 60).

```python
import seaborn as sns
plt.figure(figsize=(8,6))

sns.scatterplot(data=df, x='GDP', y='Life expectancy', hue='Status')

plt.title("GDP vs Life Expectancy")
plt.xlabel("GDP")
plt.ylabel("Life Expectancy")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns
num_cols = df.select_dtypes(include=['number'])

plt.figure(figsize=(18, 8))

sns.boxplot(data=num_cols)

plt.title("Boxplot of All Numerical Features")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python

Q1 = df['Population'].quantile(0.25)
Q3 = df['Population'].quantile(0.75)
IQR = Q3 - Q1

# Calculate lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = df[(df['Population'] < lower_bound) | (df['Population'] > upper_bound)]

# Display outliers
print("Number of outliers:", len(outliers))
print(outliers[['Country', 'Year', 'Population']])
```

```python
df[['Country', 'Population']].sort_values(by='Population', ascending=False).head(20)
```

Although the Population feature contains extreme values according to the IQR method, these values correspond to countries with genuinely large populations (e.g., India and China). Since they represent valid observations rather than data errors, the outliers were retained to preserve the integrity of the dataset

```python
corr = df.corr(numeric_only=True)['Life expectancy']

corr.sort_values().plot(kind='barh')
```

```python
plt.figure(figsize=(9, 7))
cols = ["Life expectancy", "Adult Mortality", "Alcohol", "BMI", "HIV/AIDS", "GDP", "Income composition of resources", "Schooling"]
cols = [c for c in cols if c in df.columns]
corr = df[cols].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0, linewidths=0.5)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig("7_heatmap.png")
plt.show()
```

Conclusion: Life expectancy shows a strong positive correlation with Schooling (0.74) and Income Composition of Resources (0.75), while having a strong negative correlation with Adult Mortality (-0.69) and HIV/AIDS (-0.58). This suggests that education and economic stability are the primary drivers of longevity
